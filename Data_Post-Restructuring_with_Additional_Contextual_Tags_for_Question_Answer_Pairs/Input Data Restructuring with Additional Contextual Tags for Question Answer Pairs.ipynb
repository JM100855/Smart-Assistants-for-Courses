```python
import json
import re
```

```python
def clean_tag(text):
    return re.sub(r"^[\d._]+", "", text.strip().lower().replace(" ", "_"))
```

```python
# Add tags based on heading and subheading to the extarcted text
def assign_tags(data):
    current_tags = ["introduction"]  # Starting with an "Introduction" tag until the first head/ subheading is identified
    tagged_data = []

    for entry in data:
        if "text_level" in entry:
            current_tags = [clean_tag(entry["text"])]
        else:
            entry["tags"] = current_tags
            tagged_data.append(entry)

    return tagged_data
```

```python

output_folder = "/content/drive/MyDrive/DeepLearning_FinalProject/Final Report Content/Data_Post-Restructuring_with_Additional_Contextual_Tags_for_Question_Answer_Pairs/"
for i in range(1,7):
  input_file = f"/content/drive/MyDrive/DeepLearning_FinalProject/mineru_extract-kit/chapter1-6/MIT6390Notes_Chapter{i}/auto/MIT6390Notes_Chapter{i}_content_list.json".format(i,i)
  with open(input_file, "r") as file:
      json_data = json.load(file)

  tagged_data = assign_tags(json_data)
  output_file = output_folder + f"ContextualTags_MIT6390Notes_Chapter{i}"
  with open(output_file, "w") as file:
      json.dump(tagged_data, file, indent=4)

  print(f"Tagged data saved to {output_file}")
```
