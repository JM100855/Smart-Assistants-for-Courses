```python
import json
import os
from google.colab import files
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
input_file = '/content/drive/MyDrive/merged_file (10).json'
with open(input_file, 'r') as f:
    data = json.load(f)

training_data = []
validation_data = []

for i, qa_pair in enumerate(data):
    if i % 4 < 3:
        training_data.append(qa_pair)
    else:
        validation_data.append(qa_pair)

ntraining_data = "ntraining_data.json"
with open(ntraining_data, 'w') as f:
    json.dump(training_data, f, indent=4)

nvalidation_data = "nvalidation_data.json"
with open(nvalidation_data, 'w') as f:
    json.dump(validation_data, f, indent=4)

files.download(ntraining_data)
files.download(nvalidation_data)

print(f"Both training and validation files downloaded'")
```
